top 5 country effected by mental health

```python
effected=df.groupby('COUNTRY')['MENTAL_HEALTH_AFFECT'].count().reset_index().sort_values(by='MENTAL_HEALTH_AFFECT',ascending=False).head(5)
effected
```

```python
sns.barplot(data=effected, x='COUNTRY', y='MENTAL_HEALTH_AFFECT', palette='viridis')
plt.title('Top 5 Countries Most Affected by Mental Health Issues')
plt.xlabel('Country')
plt.ylabel('Count of Mental Health Affected Cases')
plt.xticks(rotation=45)

plt.show()
```

**effected patient from particular country**

```python
df1=df.groupby(['COUNTRY','GENDER'])['MENTAL_HEALTH_AFFECT'].value_counts().reset_index().sort_values(by='MENTAL_HEALTH_AFFECT',ascending=False)
df1
```


**effected patient from particular country**


```python
df2=df1[df1['MENTAL_HEALTH_AFFECT']=='Yes']
df2
```

```python
sns.barplot(data=df2, x="COUNTRY", y="count",hue="GENDER",palette=['red','green','blue'])
plt.show()
```

**percentages of mele and female**

```python
A=df2.groupby('GENDER').size().plot(kind='pie', title=' Gender Ratio', autopct='%1.1f%%')
A
```

```python
sns.displot(data=df2, x="count",
            kind="hist",
            hue="GENDER",
            row="MENTAL_HEALTH_AFFECT",
            stat="percent",
            bins=df2["COUNTRY"].nunique()
           )

plt.show()
```
